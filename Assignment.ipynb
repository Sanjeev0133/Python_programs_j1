```python
##1.write a python program to check whether a number is positive negative or zero##
num = float(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
##2.write a python program to check if a number is even or odd##
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
##3. write a python program to find the greatest of two numbers using if else##
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print(a, "is greater.")
else:
    print(b, "is greater.")
```

```python
##5.write a python program to check if year is a leap year or not##
year = int(input("Enter a year: "))

if (year % 400 == 0):
    print(year, "is a leap year.")
elif (year % 100 == 0):
    print(year, "is not a leap year.")
elif (year % 4 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
##iterations
1.write a python program to print numbers from 1-10 using for loop
2.write a python program to calculate the sum of the first 20 natural numbers using a loop
3. write a python program to print all even numbers between 1 to 50
4.write a python program to find factorial of a number using a loop
5. write a python program to count the number of digits in a given number using while loop
```

```python
##1.write a python program to print numbers from 1-10 using for loop##
for i in range(1, 11):
    print(i)
```

```python
##2.write a python program to calculate the sum of the first 20 natural numbers using a loop##
total = 0

for i in range(1, 21):
    total += i

print("Sum of first 20 natural numbers is:", total)
```

```python
##3. write a python program to print all even numbers between 1 to 50##
for i in range(1, 51):
    if i % 2 == 0:
        print(i)
```

```python
##4.write a python program to find factorial of a number using a loop##
num = int(input("Enter a number: "))
factorial = 1

for i in range(1, num + 1):
    factorial *= i

print("Factorial of", num, "is", factorial)
```

```python
##5. write a python program to count the number of digits in a given number using while loop##
num = int(input("Enter a number: "))
count = 0
if num == 0:
    count = 1
else:
    while num != 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```
