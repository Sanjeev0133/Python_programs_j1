````python
# write a program to find the sum of all numbers from 1 to 100 using while loop
sum = 0
num = 1
while num <= 100:
    sum += num
    num += 1
print("The sum of all numbers from 1 to 100 is:", sum)
````

````python
# write a program to find sum of all even numbers between 1 to 50 using while loop
sum = 0
num = 1
while num <= 50:
    if num % 2 == 0:
        sum += num
    num += 1
print("The sum of all even numbers between 1 to 50 is:", sum)
````

````python
# write a program to find the sum of all odd numbers between 1 to 50 using while loop
sum = 0
num = 1
while num <= 50:
    if num % 2 != 0:
        sum += num
    num += 1
print("The sum of all odd numbers between 1 to 50 is:", sum)
````
